### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
locations = cities_df[['Lat', 'Lng']]
humidity = cities_df['Humidity']

atlanta_lat_ng = [33.7490, 84.3880]

fig = gmaps.figure(zoom_level=1.5, center=atlanta_lat_ng)
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 2)
fig.add_layer(heat_layer)
fig
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
good_weather_df['Hotel'] = ""
good_weather_df['Hotel_Lat'] = ""
good_weather_df['Hotel_Lng'] = ""
good_weather_df
```

```python
params = {
    "distance": 5000,
    "rankby" : "distance",
    "type":"lodging",
    "key" : gkey
}
hotel_df = good_weather_df.copy(deep=True)

for index, row in hotel_df.iterrows():
    lat = row['Lat']
    lng = row['Lng']
    params["location"] = f"{lat}, {lng}"
    url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    response = requests.get(url, params=params).json()
    results = response['results']

    try:
        hotel_df.loc[index, 'Hotel'] = results[0]['name']
        hotel_df.loc[index, 'Hotel_Lat'] = results[0]['geometry']['location']['lat']
        hotel_df.loc[index, 'Hotel_Lng'] = results[0]['geometry']['location']['lng']

    except: # catch *all* exceptions
        e = sys.exc_info()[0]
        write_to_page( "<p>Error: %s</p>" % e )
    # pprint(results)
            
print(hotel_df.shape)
hotel_df
```
